# Load employee.csv into df

```python

df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
    header=True,inferSchema=True,sep="|",quote="'"
) 
df.display()
```

## Select Operations

```python
df.select("id","name").display()
```

```python
from pyspark.sql.functions import col
df.select(col("id"),col("name")).display()
```

```python
df.select(col("id").alias("User Id"),col("name")).display()
```

## Filter Operations

```python
df.filter(col("desig")=="Developer").display()
```

```python
df.filter((col("desig")=="Developer")|(col("desig")=="Team Lead")).display()
```

```python
df.filter((col("desig")=="Developer")&(col("gen")=="M")).display()
```

```python
df.filter(col("exp").between(1,6)).display()
```
