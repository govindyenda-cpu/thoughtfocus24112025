```python

df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
    header=True,inferSchema=True,sep="|",quote="'"
) 
# df.display()
```

## Dropping missing records

```python
# df.na.drop().display()
df.na.drop(how="all").display()
```

```python
df.na.drop(subset=["name","id"]).display()
```

```python
from pyspark.sql.functions import col
# df.filter((col("id").isNotNull()) | (col("name").isNotNull())).display()
df.na.fill(0).na.fill("NULL IN SOURCE").display() 
```

```python
df.na.fill({
    "id":-1,
    "exp":0,
    "gen":"NA"
}).display()
```

```python
from pyspark.sql.functions import avg
mean = df.select(avg("exp")).collect()[0][0]
df.na.fill({
    "id":-1,
    "exp":int(mean),
    "gen":"NA"
}).display()
```
